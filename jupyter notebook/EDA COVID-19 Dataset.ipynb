```python
# File path for box directory
box_file_dir = joinpath(pwd(), "..", "..", "Box")
file_path_csv = joinpath(box_file_dir, "covid_pts_enc_level_labs_dx_2020-12-09_deid.csv")
```

```python
# Convert date features
data.admit_dt_tm = Date.(data.admit_dt_tm, "yyyy-mm-dd H:M:S")
data.discharge_dt_tm = Date.(data.admit_dt_tm, "yyyy-mm-dd H:M:S")

```

```python
# convert vent_first order
myDate(x) = Date(x, "yyyy-mm-dd H:M:S")
Date_accepting_missing = passmissing(myDate)
data.discharge_dt_tm = Date_accepting_missing.(data.discharge_dt_tm)
```

```python
# convert deceased_dt_tm
myDate(x) = Date(x, "m/dd/yyyy H:M")
Date_accepting_missing = passmissing(myDate)
data.deceased_dt_tm = Date_accepting_missing.(data.deceased_dt_tm)
```

```python
# convert vent_first order
myDate(x) = Date(x, "yyyy-mm-dd H:M:S")
Date_accepting_missing = passmissing(myDate)
data.vent_first_order_start_time = Date_accepting_missing.(data.vent_first_order_start_time)
```

```python
# convert vent_first order
myDate(x) = Date(x, "yyyy-mm-dd H:M:S")
Date_accepting_missing = passmissing(myDate)
data.vent_last_order_end_time = Date_accepting_missing.(data.vent_last_order_end_time)
```

```python
# convert icu first begin and last time
myDate(x) = Date(x, "yyyy-mm-dd H:M:S")
Date_accepting_missing = passmissing(myDate)
data.icu_first_begin_time = Date_accepting_missing.(data.icu_first_begin_time)

myDate(x) = Date(x, "yyyy-mm-dd H:M:S")
Date_accepting_missing = passmissing(myDate)
data.icu_last_end_time = Date_accepting_missing.(data.icu_last_end_time)
```

```python
# there are 303 patients who are on ventilators
filtered_vent_patients = vent_patients[completecases(vent_patients), :]
```

```python
# It appears that there are 298 unique patients. This means
# that 5 patients re admitted 
# we will use the 298 for simplicity sake
filtered_unique_vent_df = unique(filtered_vent_patients, :"deid_empi_encounter")
```

```python
filtered_unique_icu_df = unique(filtered_icu_patients, :"deid_empi_encounter")
```

```python
set_ids_vent = Set(filtered_unique_vent_df[:, "deid_empi_encounter"])
```

```python
set_ids_icu = Set(filtered_unique_icu_df[:, "deid_empi_encounter"])
```

```python
# 8 patients were not icu patients, yet they were placed on vents
intersect(set_ids_icu, set_ids_vent)
```
