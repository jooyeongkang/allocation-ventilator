```python
def sig_testing(a_df):
    
    out_list = []
    p_value_list = []

    race_list = ['Hispanic', 'White', 'Asian', 'Black or African American']
    
    combo_list = list(combinations(race_list,2))
    
    print("The group means are as follows: ")
    print('Hispanic:', a_df['death_percent'][a_df['race'] == 'Hispanic'].mean())
    print('White:', a_df['death_percent'][a_df['race'] == 'White'].mean())
    print('Black:', a_df['death_percent'][a_df['race'] == 'Black or African American'].mean())
    print('Asian:', a_df['death_percent'][a_df['race'] == 'Asian'].mean())
    print()
    
    print("The group descriptions are as follows: ")
    print('Hispanic:', a_df['death_percent'][a_df['race'] == 'Hispanic'].describe())
    print('White:', a_df['death_percent'][a_df['race'] == 'White'].describe())
    print('Black:', a_df['death_percent'][a_df['race'] == 'Black or African American'].describe())
    print('Asian:', a_df['death_percent'][a_df['race'] == 'Asian'].describe())
    print()
    
    
    print(stats.kruskal(a_df['death_percent'][a_df['race'] == 'Hispanic'],
                   a_df['death_percent'][a_df['race'] == 'White'],
                   a_df['death_percent'][a_df['race'] == 'Black or African American'],
                   a_df['death_percent'][a_df['race'] == 'Asian'],     
                   )) 
    
    for tups in combo_list:
        race_one, race_two = tups 
        race_one_dist = a_df['death_percent'][a_df['race'] == race_one]
        race_two_dist = a_df['death_percent'][a_df['race'] == race_two]
        
        test_out, p_value = stats.ks_2samp(race_one_dist, race_two_dist)
        p_value_list.append(p_value)
        if p_value <= 0.05:
            print('The distributions for {} and {} are significantly different (P = {})'.format(race_one, race_two, p_value))
            out_list.append(tups)
            
        else:
            print('The distributions for {} and {} are not significantly different (P = {})'.format(race_one, race_two, p_value))
            
            
    
    _, p_values, _, _= bonferroni(p_value_list)
    print('final p vals: ', p_values)
            
    
```

```python
def diff_scarcity1_means(all_df, type_allocation):
    thresholds = np.linspace(0,.9, 10)
    race_list = ['Hispanic', 'White', 'Asian', 'Black or African American']
    row_dict_list = []
    
    for race in race_list:
        row_dict = {}
        
        one_df = all_df[all_df['threshold'] == 1.0]
        mean_val_scarcity_one = one_df['death_percent'][one_df['race'] == race].mean()
        row_dict = {
            'Allocation': type_allocation,
            'Race': race,

        }
        row_dict[1.0] = mean_val_scarcity_one
        
        #print('For {}, the mean for 1.0 is: {}'.format(race, mean_val_scarcity_one))
        for thresh in thresholds:
            row_dict
            if thresh == 0.0:
                one_df = all_df[all_df['threshold'] == 0.0]
                mean_val_scarcity_at_zero = one_df['death_percent'][one_df['race'] == race].mean()
                #print('For {}, the mean for 0.0 is: {}'.format(race, mean_val_scarcity_at_zero))
                row_dict[thresh] = mean_val_scarcity_at_zero
            else:
                a_df = all_df[all_df['threshold'] == thresh.round(1)]
                one_df = all_df[all_df['threshold'] == 1.0]

                mean_val = a_df['death_percent'][a_df['race'] == race].mean()

                abs_diff = abs(mean_val - mean_val_scarcity_one)
                #print('For {}, the abs mean diff between 1.0 and {} is: {}'.format(race, thresh, abs_diff))
                
                row_dict['1-{}'.format(round(thresh,1))] = abs_diff
            
        row_dict_list.append(row_dict)
        #print()
        #print()
        
    return row_dict_list
        
```

```python
def mean_df(a_df, type_allocation):
    #thresholds = np.linspace(0,.9,10)
    thresholds = [.1, .2,.3,.4,.5,.6,.7,.8,.9]
    race_list = ['Hispanic', 'White', 'Asian', 'Black or African American']
    row_dict_list = []
    
    row_dict = {
            'Allocation': type_allocation,
        }
    
    a_df = a_df[a_df['threshold'].isin(thresholds)]
    
    row_dict['Hispanic'] = round(a_df['death_percent'][a_df['race'] == 'Hispanic'].mean(), 2)
    row_dict['White'] = round(a_df['death_percent'][a_df['race'] == 'White'].mean(), 2)
    row_dict['Black'] = round(a_df['death_percent'][a_df['race'] == 'Black or African American'].mean(), 2)
    row_dict['Asian'] = round(a_df['death_percent'][a_df['race'] == 'Asian'].mean(), 2)
    
    row_dict_list.append(row_dict)
    #print()
    #print()
        
    return row_dict_list
```

```python
def sig_testing_threshold(all_df):
    
    out_list = []
    race_list = ['Hispanic', 'White', 'Asian', 'Black or African American']
    
    thresholds = np.linspace(0,1,11)
    
    combo_list = list(combinations(race_list,2))
    
    for thresh in thresholds:
        p_value_list = []
        print()
        print('THRESOLD: ', thresh.round(1))
        a_df = all_df[all_df['threshold'] == thresh.round(1)]
        print("The group means are as follows: ")
        print('Hispanic:', a_df['death_percent'][a_df['race'] == 'Hispanic'].mean())
        print('White:', a_df['death_percent'][a_df['race'] == 'White'].mean())
        print('Black:', a_df['death_percent'][a_df['race'] == 'Black or African American'].mean())
        print('Asian:', a_df['death_percent'][a_df['race'] == 'Asian'].mean())
        print()

        '''
        print("The group descriptions are as follows: ")
        print('Hispanic:', a_df['death_percent'][a_df['race'] == 'Hispanic'].describe())
        print('White:', a_df['death_percent'][a_df['race'] == 'White'].describe())
        print('Black:', a_df['death_percent'][a_df['race'] == 'Black or African American'].describe())
        print('Asian:', a_df['death_percent'][a_df['race'] == 'Asian'].describe())
        print()
        '''


        print(stats.kruskal(a_df['death_percent'][a_df['race'] == 'Hispanic'],
                       a_df['death_percent'][a_df['race'] == 'White'],
                       a_df['death_percent'][a_df['race'] == 'Black or African American'],
                       a_df['death_percent'][a_df['race'] == 'Asian'],     
                       )) 

        for tups in combo_list:
            race_one, race_two = tups 
            race_one_dist = a_df['death_percent'][a_df['race'] == race_one]
            race_two_dist = a_df['death_percent'][a_df['race'] == race_two]

            test_out, p_value = stats.ks_2samp(race_one_dist, race_two_dist)
            p_value_list.append(p_value)
           

            if p_value <= 0.05:
                print('The distributions for {} and {} are significantly different (P = {})'.format(race_one, race_two, p_value))
                out_list.append(tups)

            else:
                print('The distributions for {} and {} are not significantly different (P = {})'.format(race_one, race_two, p_value))
                
        print()
        
        _, p_values, _, _= bonferroni(p_value_list)
        print('final p vals (bonferroni correction): ', p_values)
            
```

```python
def normal_plots(a_df, scheme):
    sn.distplot(a_df['death_percent'][a_df['race'] == 'Hispanic'])
    dist = plt.title('Mortality Rate for Hispanic Patients in the {} Scheme'.format(scheme))
    plt.xlabel('Mortality Rate')
    plt.show()
    
    sn.distplot(a_df['death_percent'][a_df['race'] == 'Black or African American'])
    dist = plt.title('Mortality Rate for Black Patients in the {} Scheme'.format(scheme))
    plt.xlabel('Mortality Rate')
    plt.show()
    
    sn.distplot(a_df['death_percent'][a_df['race'] == 'Asian'])
    dist = plt.title('Mortality Rate for Asian Patients in the {} Scheme'.format(scheme))
    plt.xlabel('Mortality Rate')
    plt.show()
    
    sn.distplot(a_df['death_percent'][a_df['race'] == 'White'])
    dist = plt.title('Mortality Rate for White Patients in the {} Scheme'.format(scheme))
    plt.xlabel('Mortality Rate')
    plt.show()
```
